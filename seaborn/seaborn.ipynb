# Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

v=sns.load_dataset("penguins")
v.head(3)
```

```python
# sns.barplot(x=v.island, y=v.body_mass_g)
sns.barplot(x="island", y="body_mass_g",data=v)
plt.show()
```

```python
# hue=" " => create a separate graph a/c to the requirement 
sns.barplot(x="island", y="body_mass_g",data=v,hue="sex")
plt.show()
```

```python
# order => creating a order for the x axis 
order_var=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="body_mass_g",data=v,hue="sex",order=order_var)
plt.show()
```

```python
#hue_order => hue order phle kon ayega nd last kon
sns.barplot(x="island", y="body_mass_g",data=v,hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
# ci and n_boot => size related upr black gap h jo
sns.barplot(x="island", y="body_mass_g",data=v,hue="sex",n_boot=20)
# sns.barplot(x="island", y="body_mass_g",data=v,hue="sex",ci=99)
plt.show()
```

```python
# orient="h" => creating a horizontal bar graph
sns.barplot(x="bill_depth_mm", y="bill_length_mm",data=v,hue="sex",orient="h")
plt.show()
```

```python
# palette=" " => changing the color of bar graph
# saturation=" " => changing the internsity of the color in graph
# sns.barplot(x="island", y="bill_length_mm",data=v,hue="sex",palette="icefire")
sns.barplot(x="island", y="bill_length_mm",data=v,hue="sex",palette="icefire",saturation=0.4)
plt.show()
```

```python
# errcolor errwidth capsize
sns.barplot(x="island", y="bill_length_mm",data=v,hue="sex",errcolor='y',errwidth=3,capsize=0.05)
plt.show()
```

```python
#
sns.set(style="darkgrid")
sns.barplot(x="island", y="bill_length_mm",data=v,hue="sex",alpha=1)
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

v=sns.load_dataset('penguins')
v.head(3)
```

```python
sns.displot(v['flipper_length_mm'])
plt.show()
```

```python
#bin => organise the data in a fix range
sns.displot(v['flipper_length_mm'],bins=[170,180,190,200,210,220,230])
plt.show()
```

```python
#kde=Ture => {Kernal Density Estimater} bar graph ke upr ek line milega
sns.displot(v['flipper_length_mm'],bins=[170,180,190,200,210,220,230],kde=True)
plt.show()
```

```python
# rug=True => density btyegaa graph ka niche me hii
sns.displot(v['flipper_length_mm'],bins=[170,180,190,200,210,220,230],kde=True,rug=True)
plt.show()
```

```python
# log_scale=True => log me convert kr degaa
sns.displot(v['flipper_length_mm'],log_scale=True,kde=True,rug=True)
plt.show()
```

# Strip Plot

# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

v=sns.load_dataset("tips")
v
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=v)
plt.show()
```

```python
# hue => creating  diff graph for each of the entity
#  palette => coloring diff graph with diff color
# order => changing the order of the x-axis
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=v,hue="sex",palette='dark:w',order=["Sun","Sat","Fri","Thur"])
plt.show()
```

```python
# color => coloring the graph
#  showmeans => showing the mean point in the graph
# meanprops => add all the property of marker showing the mean 
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=v,color="w",showmeans=True,meanprops={"marker":"+","markeredgecolor":"k"})
plt.show()
```

```python
# linewidth => all boundary width inc by this
# orient => for vertical and horizontal graph
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=v,showmeans=True,meanprops={"marker":"+","markeredgecolor":"k"},linewidth=4,palette="plasma",orient="v")
plt.show()
```

```python
# changing the graph to horizontal
sns.set(style="whitegrid")
sns.boxplot(y="day",x="total_bill",data=v,showmeans=True,meanprops={"marker":"+","markeredgecolor":"k"},linewidth=4,palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=v,showmeans=True,meanprops={"marker":"+","markeredgecolor":"k"},linewidth=4,palette="plasma",orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x=v["total_bill"]) # horizontal 
sns.boxplot(y=v["total_bill"]) # vertical
plt.show()
```

# Styling Plot in Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

v=sns.load_dataset("tips")
v
```

```python
# { white, whitegrid, dark, darkgrid, ticks }
sns.set_style("darkgrid") 
sns.barplot(x="day",y="total_bill",data=v,palette="plasma")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(x="day",y="total_bill",data=v)
sns.despine() # Remove the axis line 
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12,4)) # Scaling
sns.barplot(x="day",y="total_bill",data=v)
plt.show()
```

```python
sns.set_style("white")
sns.set_context("poster",font_scale=0.5)
sns.barplot(x="day",y="total_bill",data=v,palette="cool")
plt.show()
```

# Multiple Plot [Facet-Grid]

```python
 import matplotlib.pyplot as plt
import seaborn as sns

v=sns.load_dataset("tips")
v
```

```python
fg=sns.FacetGrid(v)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
# multi plot graph
fg=sns.FacetGrid(v,col="sex")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(v,col="sex",hue="day",palette="summer")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(v,col="sex",hue="day",)
fg.map(plt.scatter,"total_bill","tip",edgecolor="g").add_legend()
plt.show()
```
