# Histogram

```python
# LP - CONTRACTED
df = pd.DataFrame(var)
df['diff_1'] = df['LP'] - df['CONTRACTED']
# df
d1=pd.DataFrame(df['diff_1'])
# d1
sns.displot(df["diff_1"])
plt.show()
```

```python
# DP - LP
df = pd.DataFrame(var)
df['diff_2'] = df['DP'] - df['LP']
# df
d2=pd.DataFrame(df['diff_2'])
# d2
sns.displot(df["diff_2"])
plt.show()
```

```python
# Port Outgoing Rake - LP
df = pd.DataFrame(var)
df['diff_3'] = df['Port Outgoing Rake'] - df['LP']
# df
d3=pd.DataFrame(df['diff_3'])
# d3
sns.displot(df["diff_3"])
plt.show()
```

```python
# Plant Incoming Rake - LP
df = pd.DataFrame(var)
df['diff_4'] = df['Plant Incoming Rake'] - df['LP']
# df
d4=pd.DataFrame(df['diff_4'])
# d14
sns.displot(df["diff_4"])
plt.show()
```

```python
total_diff=pd.concat([d1,d2,d3,d4],axis=1)
total_diff
```

```python
# box plot of all above differences
sns.set_style("darkgrid")
plt.figure(figsize=(8,6))
sns.boxplot(data=total_diff,palette="cool")
plt.show()
```
