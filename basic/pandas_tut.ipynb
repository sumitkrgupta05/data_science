# Arithmetic Operations

```python
import pandas as pd
v=pd.DataFrame({"a":[1,2,3,4],"b":[5,6,7,8]})
print(v)
```

```python
import pandas as pd
v=pd.DataFrame({"a":[1,2,3,4],"b":[5,6,7,8]})
v["a+b"]=v["a"]+v["b"]
v["a-b"]=v["a"]-v["b"]
v["a*b"]=v["a"]*v["b"]
v["a/b"]=v["a"]/v["b"]
print(v)
```

```python
import pandas as pd
v=pd.DataFrame({"a":[10,20,30,40],"b":[15,16,17,18]})
v["a>20"]=v["a"] >20
print(v)
```

# Merge

```python
import pandas as pd
v1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
v2=pd.DataFrame({"A":[1,2,3,4],"C":[21,22,33,24]})
# pd.merge(v1,v2,on="A")
pd.merge(v1,v2,how="left",indicator=True)
```

```python
import pandas as pd
v1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
v2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,33,24]})
# pd.merge(v1,v2,on="A")
pd.merge(v1,v2,how="outer",indicator=True)
```

```python
import pandas as pd
v1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
v2=pd.DataFrame({"A":[1,2,3,4],"B":[21,22,33,24]})
pd.merge(v1,v2,left_index=True,right_index=True,suffixes=("name","id"))
```

# Join

```python
import pandas as pd
v1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
v2=pd.DataFrame({"X":[1,2,3],"Y":[21,22,23]})
v1.join(v2)
```

```python
v1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["*","#","/","<>"])
v2=pd.DataFrame({"X":[1,2,3],"Y":[21,22,23]},index=["*","#","/"])
v1.join(v2)
```

```python
v1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["*","#","/","<>"])
v2=pd.DataFrame({"X":[1,2],"Y":[21,22]},index=["*","#"])
v2.join(v1,how="outer",lsuffix="_ll",rsuffix="_rr")
```

# Group By

```python
# Grouping data according to their name 
import pandas as pd
var=pd.DataFrame({ "Name":["a","b","c","d","a","b","a","b","a","c","c","d"],
                   "S_1": [12,13,14,12,13,14,15,23,25,26,10,34],
                   "S_2": [23,24,25,26,27,28,29,30,25,34,35,56] })
# var
```

```python
var_new=var.groupby("Name")
for x,y in var_new:
    print(x)
    print(y)
    print()
```

```python
# Get a s[ecific group data
var_new.get_group("b")
```

```python
# Get min in all group
# var_new.min()
var_new.max()
```

# Pivot()

```python
import pandas as pd
v=pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
v
```

```python
v.pivot(index="days",columns="st_name")
```

```python
v.pivot(index="days",columns="st_name",values="maths")
```

```python
import pandas as pd
v=pd.DataFrame({"days":[1,1,1,1,2,2],"st_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,16,12],"maths":[17,18,19,13,14,16]})
v
```

```python
v.pivot_table(index="st_name",columns="days",aggfunc="sum")
```

```python
v.pivot_table(index="st_name",columns="days",aggfunc="sum",margins=True)
```
